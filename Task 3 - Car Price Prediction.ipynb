```python
num=df[['wheelbase','carlength', 'carwidth', 'carheight', 'curbweight',
     'enginesize', 'boreratio', 'stroke','compressionratio', 'horsepower',
     'peakrpm', 'citympg', 'highwaympg','price']]
cat=df[['symboling', 'CarName', 'fueltype', 'aspiration','doornumber',
     'carbody', 'drivewheel', 'enginelocation', 'enginetype',
     'cylindernumber', 'fuelsystem']]
```

The data is split into two dataframes - numerical and categorical data

### Removing outliers

```python
for i in num:
    print(i)
    sns.boxplot(data=df[i])
    plt.show()
```

```python
for i in num:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)

    IQR = Q3-Q1
    low = Q1-1.5*IQR
    up = Q3+1.5*IQR

    for j in df[i]:
        if j<=low:
            df=df.replace(j, low)
        if j>=up:
            df=df.replace(j, up)
```

### Encoding the categorical columns

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for i in cat:
    cat[i] = encoder.fit_transform(df[i])
```

```python
cat
```

### Scaling numerical columns

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(num)

cols = num.columns
scaled_dataset = pd.DataFrame(scaled_data, columns=cols)
```

```python
# Concatenate both encoded categorical dataframe and scaled numerical dataframe.

df = pd.concat([cat, scaled_dataset], axis=1)
```

```python
df
```

### Model building

```python
# Splitting the dataset

from sklearn.model_selection import train_test_split

x = df.drop(['price'], axis=1)
y = df['price']

X_test, X_train, y_test, y_train = train_test_split(x,y,random_state=42, test_size=0.3)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
# Building and fitting the model

from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train, y_train)

y_pred=model.predict(X_test)
```

### Evaluating the performance of the model

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Training Accuracy: ', model.score(X_train, y_train))
print('-----------------------------------------------------')
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print('-----------------------------------------------------')
print("R-squared:", r2_score(y_test, y_pred))
```

* The model has an accuracy of 93.1%.
* The mean squared error value is close to 0, which is an indicator of a good model.
* It also has an R-squared value of 0.731. This means that 73.1% of the variability observed in the target variable is explained by the regression model.
